 # **Distribuição uniforme** <br>
A distribuição mais direta que você provavelmente encontrará é a distribuição uniforme; A distribuição uniforme é um tipo de distribuição de probabilidade em que todos os resultados são igualmente prováveis. Por exemplo, a probabilidade de acertar em qualquer número se lançarmos um dado justo é 1/6. Esta é uma distribuição uniforme discreta.<br>
Dist_Uniforme.png



No entanto, nem todas as distribuições uniformes são discretas — elas também podem ser contínuas. Isso significa que há um número infinito de valores mensuráveis ​​igualmente prováveis ​​que podem assumir qualquer valor real dentro de um intervalo especificado. A função de densidade de probabilidade (PDF) de uma distribuição uniforme contínua entre a e b é a seguinte:<br>
Dist_Continua.png <br>
A saída do código acima: Distribuições uniformes discretas e contínuas em Python

```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy import stats

# for continuous 
a = 0
b = 50
size = 5000

X_continuous = np.linspace(a, b, size)
continuous_uniform = stats.uniform(loc=a, scale=b)
continuous_uniform_pdf = continuous_uniform.pdf(X_continuous)

# for discrete
X_discrete = np.arange(1, 7)
discrete_uniform = stats.randint(1, 7)
discrete_uniform_pmf = discrete_uniform.pmf(X_discrete) 

# plot both tables
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,5))
# discrete plot
ax[0].bar(X_discrete, discrete_uniform_pmf)
ax[0].set_xlabel("X")
ax[0].set_ylabel("Probability")
ax[0].set_title("Discrete Uniform Distribution")
# continuous plot
ax[1].plot(X_continuous, continuous_uniform_pdf)
ax[1].set_xlabel("X")
ax[1].set_ylabel("Probability")
ax[1].set_title("Continuous Uniform Distribution")
plt.show() 
```

# **Distribuição gaussiana**<br>
A distribuição gaussiana provavelmente será a distribuição com a qual você estará mais familiarizado. Tem vários nomes: alguns chamam de curva de sino, já que o gráfico de sua probabilidade se parece com um sino, alguns chamam de distribuição gaussiana em homenagem a Carl Gauss, que foi o matemático alemão que a descreveu pela primeira vez, e outros chamam de distribuição normal porque os primeiros estatísticos notou sua recorrência uma e outra vez.
A função densidade de probabilidade de uma distribuição normal é a seguinte:<br>
Dist_Gaugasiana.png<br>
σ é o desvio padrão e μ é a média da distribuição. Mas é importante notar que em uma distribuição normal, a média, moda e mediana são todas iguais.
Quando plotamos variáveis ​​aleatórias normalmente distribuídas, a curva é simétrica em torno da média – metade dos valores estão à esquerda e metade à direita do centro. E, a área total sob a curva é 1.<br>

```python
# credit: @unutbu - https://stackoverflow.com/questions/10138085/how-to-plot-normal-distribution
mu = 0
variance = 1
sigma = np.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

plt.subplots(figsize=(8, 5))
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.title("Normal Distribution")
plt.show()
```

Algo conhecido sobre a distribuição normal é a regra empírica. A regra empírica nos informa a porcentagem de dados que se enquadra em um certo número de desvios padrão da média. Essas porcentagens são:
68% dos dados estão dentro de um desvio padrão da média.
95% dos dados estão dentro de dois desvios padrão da média.
99,7% dos dados estão dentro de três desvios padrão da média.

# **Distribuição lognormal** <br>
Uma distribuição lognormal é uma distribuição de probabilidade contínua de uma variável aleatória cujo logaritmo é normalmente distribuído. Assim, se a variável aleatória X é log-normalmente distribuída, então Y = ln( X ) tem uma distribuição normal .
[ Fonte : Wikipédia ]<br>
Dist_Lognormal.png<br>
Uma variável aleatória que é lognormalmente distribuída recebe apenas valores reais positivos. Consequentemente, as distribuições lognormais criam curvas assimétricas à direita

```python
X = np.linspace(0, 6, 500)

std = 1
mean = 0
lognorm_distribution = stats.lognorm([std], loc=mean)
lognorm_distribution_pdf = lognorm_distribution.pdf(X)

fig, ax = plt.subplots(figsize=(8, 5))
plt.plot(X, lognorm_distribution_pdf, label="μ=0, σ=1")
ax.set_xticks(np.arange(min(X), max(X)))

std = 0.5
mean = 0
lognorm_distribution = stats.lognorm([std], loc=mean)
lognorm_distribution_pdf = lognorm_distribution.pdf(X)
plt.plot(X, lognorm_distribution_pdf, label="μ=0, σ=0.5")

std = 1.5
mean = 1
lognorm_distribution = stats.lognorm([std], loc=mean)
lognorm_distribution_pdf = lognorm_distribution.pdf(X)
plt.plot(X, lognorm_distribution_pdf, label="μ=1, σ=1.5")

plt.title("Lognormal Distribution")
plt.legend()
plt.show()
```

#  Distribuição de Poisson  <br>
A distribuição de Poisson recebeu o nome de um matemático francês chamado Siméon Denis Poisson . É uma distribuição de probabilidade discreta, o que significa que conta ocorrências que têm resultados finitos — em outras palavras, é uma distribuição de contagem. Assim, a distribuição de Poisson é usada para mostrar o número de vezes que um evento provavelmente ocorrerá em um período especificado.
Se um evento ocorre com uma taxa fixa no tempo, então a probabilidade de observar o número (n) de eventos no tempo pode ser descrita com a distribuição de Poisson. Por exemplo, um cliente pode chegar a um café a uma taxa média de 3 por minuto. Podemos usar a distribuição de Poisson para calcular a probabilidade de chegada de 9 clientes em 2 minutos.
Aqui está a fórmula da função de massa de probabilidade para ajudar:<br>
Dist_Poisson.png<br>
λ é a taxa de eventos em uma unidade de tempo — em nosso exemplo, é 3. k é o número de ocorrências — em nosso exemplo, é 9. Podemos usar para resolver nosso problema de exemplo. Scipy

```python
from scipy import stats

print(stats.poisson.pmf(k=9, mu=3))
"""
0.002700503931560479
"""
```

A curva de uma distribuição de Poisson é semelhante a uma distribuição normal e λ marca o pico. Vamos plotar um em Python para ver como isso fica visualmente.

```python
# generate random values from poisson distribution with sample size of 500
X = stats.poisson.rvs(mu=3, size=500)

plt.subplots(figsize=(8, 5))
plt.hist(X, density=True, edgecolor="black")
plt.title("Poisson Distribution")
plt.show()
```

# **Distribuição binomial** <br>
Você pode pensar na distribuição binomial como a probabilidade de sucesso ou fracasso em um experimento. Alguns também podem descrevê-lo como a probabilidade de lançamento da moeda.
Aqui está a definição da Wikipedia:
a distribuição binomial com parâmetros n e p é a distribuição de probabilidade discreta do número de sucessos em uma seqüência de n experimentos independentes, cada um fazendo uma pergunta sim-não e cada um com seu próprio resultado de valor booleano: sucesso ou fracasso.<br>
[ Fonte : Wikipédia ]
Essencialmente, a distribuição binomial mede a probabilidade de dois eventos. A probabilidade de um evento ocorrer com a probabilidade p e o outro evento ocorrer com a probabilidade 1-p.
Esta é a fórmula para uma distribuição binomial:<br>
Dist_Binomial.png

```python
# Binomial
X = np.random.binomial(n=1, p=0.5, size=1000)

plt.subplots(figsize=(8, 5))
plt.hist(X)
plt.title("Binomial Distribution")
plt.show()
```

# **Distribuição T de Student** <br>
A distribuição t de Student (ou simplesmente a distribuição t ) é qualquer membro de uma família de distribuições de probabilidade contínuas que surgem ao estimar a média de uma população normalmente distribuída em situações em que o tamanho da amostra é pequeno e o desvio padrão da população é desconhecido. Foi desenvolvido pelo estatístico inglês William Sealy Gosset sob o pseudônimo de “Student”.
[ Fonte : Wikipédia ] <br>
Dist_T.png
<br>
n é o parâmetro livre chamado “ graus de liberdade ”, mas você também pode vê-lo referido como “ d.o.f. ”. A distribuição t se aproxima de uma distribuição normal para valores altos de n.

```python
import seaborn as sns
from scipy import stats

X1 = stats.t.rvs(df=1, size=4)
X2 = stats.t.rvs(df=3, size=4)
X3 = stats.t.rvs(df=9, size=4)

plt.subplots(figsize=(8,5))
sns.kdeplot(X1, label = "1 d.o.f")
sns.kdeplot(X2, label = "3 d.o.f")
sns.kdeplot(X3, label = "6 d.o.f")
plt.title("Student's t distribution")
plt.legend()
plt.show()
```

# **Distribuição qui-quadrado** <br>
A distribuição qui-quadrado é um caso especial da distribuição gama; Com k graus de liberdade, a distribuição qui-quadrado é a soma dos quadrados de k para algumas variáveis ​​aleatórias normais padrão independentes.<br>
Dist_Quiquadrado.png<br>
É uma distribuição de probabilidade popular, comumente usada em testes de hipóteses e na construção de intervalos de confiança.

```python
#x-axis ranges from 0 to 20 with .001 steps
X = np.arange(0, 6, 0.25)

plt.subplots(figsize=(8, 5))
plt.plot(X, stats.chi2.pdf(X, df=1), label="1 d.o.f")
plt.plot(X, stats.chi2.pdf(X, df=2), label="2 d.o.f")
plt.plot(X, stats.chi2.pdf(X, df=3), label="3 d.o.f")
plt.title("Chi-squared Distribution")
plt.legend()
plt.show()
```
